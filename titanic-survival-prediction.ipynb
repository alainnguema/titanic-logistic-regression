```python
categories_ticket = df['Ticket'].unique()

def transformation(value):
    value_r = "numeric" if value.isnumeric() else value
    value_r = value_r.split(' ')[0]
    value_r = value_r.replace('.','').replace('/','').upper()[:2]
    return value_r

categories_2 = list(set([transformation(ele) for ele in categories_ticket]))
print(len(categories_ticket), len(categories_2))
print(categories_2)
```

```python
df = pd.get_dummies(df, columns=['Sex','Ticket','Embarked'])
df.info()
```

```python
numpy_dataset = df.values
print(numpy_dataset.shape)
```

```python
X, Y = numpy_dataset[:,1:], numpy_dataset[:, 0:1]
print(X.shape, Y.shape)
```

```python
def normalization(X):
    """
    Normalise uniquement les colonnes numériques (non-booléennes).
    Les colonnes booléennes restent inchangées.
    """
    X_normalized = X.copy().astype(float)
    _min = np.min(X, axis=0)
    _max = np.max(X, axis=0)
    
    # Éviter la division par zéro pour les colonnes constantes
    # Normaliser uniquement les colonnes où max != min
    mask = (_max - _min) != 0
    X_normalized[:, mask] = (X[:, mask] - _min[mask]) / (_max[mask] - _min[mask])
    
    return X_normalized, _min, _max

new_X, _min, _max = normalization(X)
print(_min, _max)
```

```python
new_X
```

```python
def hypothesis(X, theta):
    X_ = np.concatenate([np.ones((X.shape[0], 1)), X], axis=1)
    Z = np.dot(X_, theta)
    Z = np.array(Z, dtype=np.float64)
    return 1 / (1 + np.exp(-Z))

theta_test = np.random.random((new_X.shape[1] + 1, 1))
print(hypothesis(X[:2,:], theta_test))
```

```python
def cost(X, Y, theta):
    H = hypothesis(X, theta)
    Z = Y*np.log(H)+(1-Y)*np.log(1-H)
    return (-1/(X.shape[0]))*np.sum(Z)

print(cost(new_X, Y, theta_test))
```

```python
def gradient(X,Y,theta):
    H = hypothesis(X,theta)
    X_ = np.concatenate([np.ones((X.shape[0], 1)), X], axis=1)
    return (1/X.shape[0]) * np.dot(X_.T, H-Y)

print(gradient(new_X, Y, theta_test).shape)
```

```python
def accuracy(X,Y,theta):
    H = hypothesis(X, theta)
    pred = H > 0.5
    nb_good_prediction = np.sum(pred==Y)
    return 100*(nb_good_prediction/X.shape[0])

def gradient_descent(X,Y,lr, steps):
    theta_i = np.random.random((X.shape[1]+1,1))
    for i in range(steps):
        grad = gradient(X,Y,theta_i)
        theta_i = theta_i - lr*grad
        if i%10000 == 9999:
            print('Cost = ',cost(X,Y,theta_i), 'and accuracy = ', accuracy(X,Y,theta_i),' %')
    return theta_i

# Réduire le nombre d'itérations pour un test plus rapide (vous pouvez augmenter si nécessaire)
theta_r = gradient_descent(new_X,Y,1e-3,100000)
```
